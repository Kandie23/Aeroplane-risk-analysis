# Aircraft Purchase Risk Assessment
## Project overview

Using structured data on aircraft specifications, operational history, maintenance records, and financial metrics, we will develop a risk evaluation framework. This framework will help identify aircrafts that pose high operational or financial risks before acquisition.

## Objectives

- To understand the structure, quality, and completeness of the aircraft dataset.
- To explore patterns and relationships between aircraft attributes and risk indicators.
- To engineer new features that reflect usage intensity, depreciation, and maintenance burden.
- To build a predictive model (classification or clustering) to categorize aircraft by risk level.
- To provide actionable insights for procurement strategy based on historical data trends.

### Injury Data Imputation 

This code handles missing values in injury columns by intelligently leveraging information from the `Injury.Severity` text field. The approach ensures accurate imputation while prioritizing safety-critical fatal injury data.


```python
# Injury counts - fill missing with 0 (no injuries)
injury_cols = ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 
               'Total.Minor.Injuries', 'Total.Uninjured']
df[injury_cols] = df[injury_cols].fillna(0)

# Injury Severity - derive from injury counts
df['Injury.Severity'] = np.select(
    [
        df['Total.Fatal.Injuries'] > 0,
        (df['Total.Serious.Injuries'] > 0) | (df['Total.Minor.Injuries'] > 0)
    ],
    ['Fatal', 'Non-Fatal'],
    default='Non-Fatal'
)

# Aircraft damage - impute based on injury severity
damage_map = {
    'Fatal': 'Destroyed',
    'Non-Fatal': 'Substantial'
}
df['Aircraft.damage'] = df['Aircraft.damage'].fillna(
    df['Injury.Severity'].map(damage_map)
)
```

```python
df.isna().sum()
```

```python
df['Injury.Severity'].value_counts()
```

#### Handling the remaining missing values

```python
# Aircraft category - high missingness domain
df['Aircraft.Category'] = df['Aircraft.Category'].fillna('Unknown')

# Engine specifications - preserve "unknown" status
df['Engine.Type'] = df['Engine.Type'].fillna('Uknown')

# Engine count with median imputation
engine_median = round(df['Number.of.Engines'].median())
df['Number.of.Engines'] = df['Number.of.Engines'].fillna(engine_median).astype(int)

# Weather with UNK
df['Weather.Condition'] = df['Weather.Condition'].fillna('Unk')

# Final validation
assert df.isna().sum().sum() == 0
print("Missing values handled successfully!")
```

### Risk feature engineering

```python
# Total occupants onboard
df['Total_Occupants'] = (df['Total.Fatal.Injuries'] + 
                         df['Total.Serious.Injuries'] + 
                         df['Total.Minor.Injuries'] + 
                         df['Total.Uninjured'])

# Fatality rate 
df['Fatality_Rate'] = np.where(
    df['Total_Occupants'] > 0,
    df['Total.Fatal.Injuries'] / df['Total_Occupants'],
    0
)

# Binary risk indicator
df['High_Risk_Flag'] = np.where(
    (df['Aircraft.damage'] == 'Destroyed') | 
    (df['Injury.Severity'] == 'Fatal'),
    1, 0
)

# Weather risk score
weather_risk = {
    'VMC': 1,   # Visual conditions = low risk
    'IMC': 3,   # Instrument conditions = high risk
    'UNK': 2,   # Unknown = medium risk
    '': 2       # Missing = medium risk
}
df['Weather_Risk'] = df['Weather.Condition'].map(weather_risk)
```

##  EDA

We'll conduct a comprehensive EDA to uncover patterns and relationships in the aircraft risk data. The analysis will focus on risk distribution, aircraft characteristics, and safety metrics.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Risk Distribution Analysis

plt.figure(figsize=(8, 5))
risk_counts = df1['High_Risk_Flag'].value_counts(normalize=True) * 100
palette = sns.color_palette("viridis", 2)
ax = sns.barplot(x=risk_counts.index, y=risk_counts.values, palette=palette)
plt.title('Distribution of High-Risk Incidents', fontsize=14)
plt.xlabel('Risk Classification', fontsize=11)
plt.ylabel('Percentage (%)', fontsize=11)
plt.xticks([0, 1], ['Low Risk (0)', 'High Risk (1)'])
```

The bar graphs indicate that 30% of incidents are high-risk  a critical minority that could drive to losses if procurement decision is not done with some thought in it.

```python
 # Aircraft Damage Analysis
    
plt.figure(figsize=(10, 6))
damage_counts = df1['Aircraft.damage'].value_counts()
palette = sns.color_palette("mako_r", len(damage_counts))
ax = sns.barplot(x=damage_counts.values, y=damage_counts.index, palette=palette, orient='h')
plt.title('Aircraft Damage Severity', fontsize=14)
plt.xlabel('Count', fontsize=11)
plt.ylabel('Damage Category', fontsize=11)
```

Substantial damage- Most common, indicating frequent high-cost repairs

Destroyed aircraft - Fewer occurrences but catastrophic losses

Minor damage - Lower financial impact

```python
# Manufacturer Incident Analysis
    
plt.figure(figsize=(10, 7))
top_makes = df1['Make'].value_counts().nlargest(10)
palette = sns.color_palette("rocket_r", len(top_makes))
ax = sns.barplot(y=top_makes.index, x=top_makes.values, palette=palette, orient='h')
plt.title('Top 10 Manufacturers by Incident Count', fontsize=14)
plt.xlabel('Number of Incidents', fontsize=11)
plt.ylabel('Manufacturer', fontsize=11)
```

```python
# Manufacturer Risk Profile
plt.figure(figsize=(10, 7))
make_risk = df1.groupby('Make')['High_Risk_Flag'].mean().sort_values(ascending=False) * 100
top_risky = make_risk[make_risk > 0].nlargest(10)
palette = sns.color_palette("magma_r", len(top_risky))
ax = sns.barplot(y=top_risky.index, x=top_risky.values, palette=palette, orient='h')
plt.title('Top 10 Riskiest Manufacturers', fontsize=14)
plt.xlabel('High-Risk Rate (%)', fontsize=11)
plt.ylabel('Manufacturer', fontsize=11)
plt.xlim(0, 60)
```

```python
# 7. Aircraft Category Analysis
plt.figure(figsize=(10, 6))
category_risk = df1.groupby('Aircraft.Category')['High_Risk_Flag'].mean().sort_values(ascending=False) * 100
palette = sns.color_palette("viridis", len(category_risk))
ax = sns.barplot(y=category_risk.index, x=category_risk.values, palette=palette, orient='h')
plt.title('High-Risk Rate by Aircraft Category', fontsize=14)
plt.xlabel('High-Risk Rate (%)', fontsize=11)
plt.ylabel('Aircraft Category', fontsize=11)
plt.xlim(0, 50)
```

```python
#Occupancy vs Fatal Injuries
plt.figure(figsize=(10, 6))
ax = sns.scatterplot(data=df1, x='Total_Occupants', y='Total.Fatal.Injuries', 
                     hue='High_Risk_Flag', palette="viridis", alpha=0.7, size='High_Risk_Flag', sizes=[20, 60])
plt.title('Fatal Injuries vs Total Occupants', fontsize=14)
plt.xlabel('Total Occupants', fontsize=11)
plt.ylabel('Fatal Injuries', fontsize=11)
plt.xlim(0, 50)
plt.ylim(0, 30)
plt.legend(title='Risk Level', labels=['Low Risk', 'High Risk'])
plt.tight_layout()
plt.show()
```

```python
# Calculate model risk with minimum incident threshold
model_risk = df1.groupby('Model').agg(
    Incident_Count=('High_Risk_Flag', 'count'),
    High_Risk_Rate=('High_Risk_Flag', 'mean')
).query('Incident_Count >= 100').sort_values('High_Risk_Rate', ascending=False).head(15)
```

```python
# Aggregate by make and model
model_incidents = df1.groupby(['Make', 'Model']).agg(
    Total_Incidents=('High_Risk_Flag', 'count'),
    High_Risk_Rate=('High_Risk_Flag', 'mean')
).reset_index().sort_values('Total_Incidents', ascending=False)

# Filter for high-volume models (>500 incidents)
top_models = model_incidents[model_incidents['Total_Incidents'] >= 500]
top_models = top_models.sort_values('Total_Incidents', ascending=False)
```

```python
plt.figure(figsize=(12, 8))
ax = sns.barplot(
    y=top_models['Model'], 
    x=top_models['Total_Incidents'],
    hue=top_models['Make'],
    dodge=False,
    palette=sns.color_palette("husl", len(top_models))
)

plt.title('Aircraft Models with Highest Incident Volume (500+ Incidents)', fontsize=14)
plt.xlabel('Total Incidents', fontsize=11)
plt.ylabel('Aircraft Model', fontsize=11)
plt.legend(title='Manufacturer', loc='lower right')

# Annotate with risk rates
for i, row in enumerate(top_models.itertuples()):
    ax.text(row.Total_Incidents + 50, i, 
            f"{row.High_Risk_Rate*100:.1f}% high-risk", 
            va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# Create combined Make-Model column
top_models['Aircraft'] = top_models['Make'] + ' ' + top_models['Model']

# Sort by incident count
top_models = top_models.sort_values('Total_Incidents', ascending=True)  # Reverse for horizontal bar

# Create visualization
plt.figure(figsize=(12, 8))
colors = sns.color_palette("viridis", len(top_models))
ax = sns.barplot(
    x='Total_Incidents',
    y='Aircraft',
    data=top_models,
    palette=colors
)

plt.title('Top 10 Aircraft by Incident Volume (500+ Incidents)', fontsize=14)
plt.xlabel('Total Incident Reports', fontsize=11)
plt.ylabel('Aircraft Type', fontsize=11)
plt.xlim(0, 4500)

# Annotate with risk rates and incident counts
for i, aircraft in enumerate(top_models['Aircraft']):
    row = top_models[top_models['Aircraft'] == aircraft].iloc[0]
    ax.text(
        row.Total_Incidents + 100, 
        i,
        (f"{row.Total_Incidents:,} incidents | "
         f"{row.High_Risk_Rate*100:.1f}% high-risk"),
        va='center',
        fontsize=9
    )

plt.tight_layout()
plt.savefig('Top_Aircraft_Incidents.png', dpi=120, bbox_inches='tight')
plt.show()
```

```python
# Calculate model safety with minimum 100 incidents
model_safety = df1.groupby(['Make', 'Model']).agg(
    Incident_Count=('High_Risk_Flag', 'count'),
    High_Risk_Rate=('High_Risk_Flag', 'mean')
).reset_index().query('Incident_Count >= 100')

# Get top 10 safest (lowest risk rate)
top_safest = model_safety.sort_values('High_Risk_Rate').head(10)
top_safest['Aircraft'] = top_safest['Make'] + ' ' + top_safest['Model']

# Create visualization
plt.figure(figsize=(12, 8))
colors = sns.color_palette("crest_r", len(top_safest))
ax = sns.barplot(
    y='Aircraft',
    x='High_Risk_Rate',
    data=top_safest,
    palette=colors,
    orient='h',
    order=top_safest.sort_values('High_Risk_Rate', ascending=False)['Aircraft']
)

plt.title('Top 10 Aircraft with the least incidents Models (100+ Incident Threshold)', fontsize=14)
plt.xlabel('High-Risk Rate', fontsize=11)
plt.ylabel('Aircraft Model', fontsize=11)
plt.xlim(0, 0.15)

# Annotate with safety metrics
for i, row in enumerate(top_safest.itertuples()):
    ax.text(
        row.High_Risk_Rate + 0.005, 
        i,
        (f"Risk: {row.High_Risk_Rate*100:.1f}% | "
         f"Incidents: {row.Incident_Count:,}"),
        va='center',
        fontsize=9
    )

plt.tight_layout()
plt.savefig('Top_10_Safest_Aircraft.png', dpi=120, bbox_inches='tight')
plt.show()
```
